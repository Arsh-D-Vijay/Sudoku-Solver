# Sudoku Solver

### Introduction
The aim of this project is to understand and visualise backtracking and recurssion.

A sudoku is a 9x9 square matrix divided in 9 equal squares of 3x3.

### Rules:
The number to be filled in a position should ***not*** be present in :
* Column
* Row
* Its 3x3 square

```python
board = [[3, 0, 6, 5, 0, 8, 4, 0, 0], 
         [5, 2, 0, 0, 0, 0, 0, 0, 0], 
         [0, 8, 7, 0, 0, 0, 0, 3, 1], 
         [0, 0, 3, 0, 1, 0, 0, 8, 0], 
         [9, 0, 0, 8, 6, 3, 0, 0, 5], 
         [0, 5, 0, 0, 9, 0, 6, 0, 0], 
         [1, 3, 0, 0, 0, 0, 2, 5, 0], 
         [0, 0, 0, 0, 0, 0, 0, 7, 4], 
         [0, 0, 5, 2, 0, 6, 3, 0, 0] ]
```

```python
def Print_Board(Board):
    for i in range(9):
        if i == 3 or i == 6:
            print("------------------------------------")
        for j in range(9):
            if j == 3 or j == 6 :
                print("| ", end='')
            print(Board[i][j], " ", end = "")
            if j==8:
                print('|')
    print("\n")
```

```python
def Find_pos(bo):
    for x in range(9):
        for y in range(9):
            if bo[x][y] == 0:
                return (x,y)
    return (-1,-1)
```

```python
def Is_valid(board,num,pos):
    for i in range(9): 
        if board[pos[0]][i] == num:
            return False
    for j in  range(9):
        if board[j][pos[1]] == num:
            return False
    bo_x = pos[0]//3
    bo_y = pos[1]//3
    for i in range(bo_x*3,bo_x*3 +3):
        for j in range(bo_y*3,bo_y*3 + 3):
            if board[i][j]==num:
                return False
    return True
```

```python
def solve(board):
    Print_Board(board)
    pos = Find_pos(board)
    if pos[0] == -1:
        return True
    for i in range(1,10):
        if Is_valid(board,i,pos) == True:
            board[pos[0]][pos[1]] = i
            if solve(board):
                return True
            board[pos[0]][pos[1]] = 0
    return False
```

```python
solve(board)
```
